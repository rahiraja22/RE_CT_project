```python
import zipfile

# Extract the uploaded zip file
with zipfile.ZipFile("HR_dataset.zip", 'r') as zip_ref:
    zip_ref.extractall("HR_dataset")
```

```python
import pandas as pd
import glob

# Path
all_files = glob.glob("HR_dataset/HR_dataset/*.csv")

print("Total CSV files found:", len(all_files))

# Read with encoding fallback
dfs = []
for f in all_files:
    try:
        df_temp = pd.read_csv(f, encoding='utf-8')
    except UnicodeDecodeError:
        df_temp = pd.read_csv(f, encoding='latin-1')  # fallback encoding
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)

print(df.shape)
print(df.head())
```

```python
df.info()
```

```python
df = df.applymap(lambda x: str(x).replace("`", "") if isinstance(x, str) else x)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
```

```python
num_cols = df.columns.drop(["india/states", "nic_name"])
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python

df["state_code"] = df["state_code"].astype(int)
df["district_code"] = df["district_code"].astype(int)
```

```python
%%writefile hr_app.py
import streamlit as st
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Load data
# -----------------------------
st.set_page_config(page_title="Workers EDA Dashboard", layout="wide")

st.title("📊 Workers Industrial Classification - EDA Dashboard")

df = pd.read_csv("clean_hr_data.csv")

# Define numeric cols globally
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# -----------------------------
# 2. Pre-compute Aggregates
# -----------------------------

# Total by State
state_total = df.groupby("india/states")[
    ["main_workers___total____persons",
     "marginal_workers___total____persons"]
].sum().reset_index()
state_total["total_workers"] = (
    state_total["main_workers___total____persons"]
    + state_total["marginal_workers___total____persons"]
)

# Total by Industry
industry_total = df.groupby("nic_name")[
    ["main_workers___total____persons",
     "marginal_workers___total____persons"]
].sum().reset_index()
industry_total["total_workers"] = (
    industry_total["main_workers___total____persons"]
    + industry_total["marginal_workers___total____persons"]
)

# State + Industry
state_industry_agg = df.groupby(["india/states", "nic_name"])[
    ["main_workers___total____persons", "marginal_workers___total____persons"]
].sum().reset_index()
state_industry_agg["total_workers"] = (
    state_industry_agg["main_workers___total____persons"]
    + state_industry_agg["marginal_workers___total____persons"]
)

# Male vs Female by State
gender_state = df.groupby("india/states")[
    ["main_workers___total___males", "main_workers___total___females",
     "marginal_workers___total___males", "marginal_workers___total___females"]
].sum().reset_index()
gender_state["male_workers"] = (
    gender_state["main_workers___total___males"] + gender_state["marginal_workers___total___males"]
)
gender_state["female_workers"] = (
    gender_state["main_workers___total___females"] + gender_state["marginal_workers___total___females"]
)

# Rural vs Urban
rural_urban = df.groupby("india/states")[
    ["main_workers___rural____persons", "main_workers___urban____persons",
     "marginal_workers___rural____persons", "marginal_workers___urban____persons"]
].sum().reset_index()
rural_urban["rural_workers"] = (
    rural_urban["main_workers___rural____persons"] + rural_urban["marginal_workers___rural____persons"]
)
rural_urban["urban_workers"] = (
    rural_urban["main_workers___urban____persons"] + rural_urban["marginal_workers___urban____persons"]
)

# -----------------------------
# 3. Sidebar Controls
# -----------------------------
st.sidebar.header("🔎 Select EDA View")
eda_option = st.sidebar.selectbox(
    "Choose a Visualization",
    [
        "Overall Stats",
        "Top States by Workers",
        "Top Industries",
        "Gender Analysis",
        "Rural vs Urban",
        "Heatmap / Correlation",
        "Distribution Histogram",
        "Outliers Detection"
    ]
)

# -----------------------------
# 4. Visualization Sections
# -----------------------------

if eda_option == "Overall Stats":
    st.subheader("📌 Overall Workers Statistics")
    total_workers = (
        df["main_workers___total____persons"].sum()
        + df["marginal_workers___total____persons"].sum()
    )
    main_workers = df["main_workers___total____persons"].sum()
    marginal_workers = df["marginal_workers___total____persons"].sum()
    male_total = (
        df["main_workers___total___males"].sum()
        + df["marginal_workers___total___males"].sum()
    )
    female_total = (
        df["main_workers___total___females"].sum()
        + df["marginal_workers___total___females"].sum()
    )

    st.metric("Total Workers", f"{total_workers:,}")
    st.metric("Main Workers", f"{main_workers:,}")
    st.metric("Marginal Workers", f"{marginal_workers:,}")
    st.metric("Male Workers", f"{male_total:,}")
    st.metric("Female Workers", f"{female_total:,}")

elif eda_option == "Top States by Workers":
    st.subheader("🏆 Top States by Total Workers")
    top_n = st.sidebar.slider("Select Top N States", 5, 20, 10)
    top_states = state_total.sort_values("total_workers", ascending=False).head(top_n)
    fig = px.bar(top_states, x="india/states", y="total_workers",
                 title=f"Top {top_n} States by Total Workers")
    st.plotly_chart(fig, use_container_width=True)

elif eda_option == "Top Industries":
    st.subheader("🏭 Top Industries by Workers")
    top_n = st.sidebar.slider("Select Top N Industries", 5, 20, 10)
    top_industries = industry_total.sort_values("total_workers", ascending=False).head(top_n)
    fig = px.bar(top_industries, x="nic_name", y="total_workers",
                 title=f"Top {top_n} Industries by Workers")
    st.plotly_chart(fig, use_container_width=True)

elif eda_option == "Gender Analysis":
        st.subheader("👩‍🦱 vs 👨 Gender Split")
        gender_totals = {
            "Males": df["main_workers___total___males"].sum(),
            "Females": df["main_workers___total___females"].sum()
        }
        fig = px.bar(x=list(gender_totals.keys()), y=list(gender_totals.values()),
                     title="Overall Gender Split", text_auto=True)
        st.plotly_chart(fig, use_container_width=True)


elif eda_option == "Rural vs Urban":
    st.subheader("🏠 Rural vs 🌆 Urban")
    rural_urban_totals = {
        "Rural": df["main_workers___rural____persons"].sum() + df["marginal_workers___rural____persons"].sum(),
        "Urban": df["main_workers___urban____persons"].sum() + df["marginal_workers___urban____persons"].sum()
    }
    fig = px.pie(names=list(rural_urban_totals.keys()), values=list(rural_urban_totals.values()),
                 title="Rural vs Urban Workers")
    st.plotly_chart(fig, use_container_width=True)


elif eda_option == "Heatmap / Correlation":
    st.subheader("🔥 Correlation Heatmap")
    corr = df[numeric_cols].corr()
    fig, ax = plt.subplots(figsize=(15, 10))
    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, ax=ax)
    st.pyplot(fig)

elif eda_option == "Distribution Histogram":
    st.subheader("📈 Distribution of Numeric Columns")
    selected_col = st.sidebar.selectbox("Choose numeric column:", numeric_cols)
    fig = px.histogram(df, x=selected_col, nbins=50,
                       title=f"Histogram of {selected_col}", marginal="box")
    st.plotly_chart(fig, use_container_width=True)


elif eda_option == "Outliers Detection":
    st.subheader("📦 Outlier Detection Across All Numeric Columns")

    # numeric columns automatically detect
    numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

    import matplotlib.pyplot as plt
    import seaborn as sns

    fig, ax = plt.subplots(figsize=(15, 8))
    sns.boxplot(data=df[numeric_cols], orient="h", ax=ax, palette="Set2", fliersize=4)
    ax.set_title("Boxplots of All Numeric Columns", fontsize=16)
    st.pyplot(fig)

```
